# Trees

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the data

```python
df = pd.read_csv('data/kyphosis.csv')
df.head()
```

```python
df.info()
```

```python
# sns.distplot(df['Age'], bins=10, kde=False)
sns.pairplot(df, hue='Kyphosis')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis', axis=1)
```

```python
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('-----------')
print(classification_report(y_test, predictions))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_predictions = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test, rfc_predictions))
print('-----------')
print(classification_report(y_test, rfc_predictions))
```

## Tree visualization

```python
from IPython.display import Image
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz

import pydot
```

## For the Decision Tree

```python
features = list(df.columns[1:])
features
```

```python
dot_data = StringIO()
```

```python
# rounded: 
# filled: 
export_graphviz(dtree, out_file=dot_data, feature_names=features, rounded=True, filled=True)
```

```python
graph = pydot.graph_from_dot_data(dot_data.getvalue())
Image(graph[0].create_png(), )
```

## For the Random Forest Classifier

```python
estimator = rfc.estimators_[10]
features = list(df.columns[1:])
classes = list(df.columns[0])
```

```python
from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(estimator, out_file='img/Random_Forest_Classifier.dot', 
                feature_names = features,
                class_names = classes,
                rounded = True, proportion = False, 
                precision = 2, filled = True)
```

```python
# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'img/Random_Forest_Classifier.dot', '-o', 'img/Random_Forest_Classifier.png', '-Gdpi=600'])
```
