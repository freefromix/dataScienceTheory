# K-means

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
# c : color, sequence, or sequence of color, optional
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.grid(True)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2)= plt.subplots(1,2, sharey=True, figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
alldata = np.stack((data[0][:,0], data[0][:,1],kmeans.labels_))
#test[numpy.logical_or.reduce()]
x = data[0][:,0]
y = data[0][:,1]
z = data[1]
alldata = np.array([x,y, z])

cluster0 = alldata[0:2,alldata[2]==0]
cluster1 = alldata[0:2,alldata[2]==1]
cluster2 = alldata[0:2,alldata[2]==2]
cluster3 = alldata[0:2,alldata[2]==3]
```
