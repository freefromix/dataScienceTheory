```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Comes from here: https://github.com/mwaskom/seaborn-data
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
setosas = iris[(iris['species']  == 'setosa')]
setosas
```

```python
setosas_mean_sepal_width = setosas['sepal_width'].mean()
maxiris = iris['sepal_width'].max()
print(setosas_mean_sepal_width, "\t", maxiris)
```

```python
# Comes from here: https://github.com/mwaskom/seaborn-data
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
# pairplot: Diagonal axis is probability of occurence (univariate distribution) 
g = sns.pairplot(iris, hue='species')
```

```python
# PairGrid: diagonal axis can be whatever you want (distance plot here)
g = sns.PairGrid(iris, hue='species')
g.map_diag(func=sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(func=sns.kdeplot)
```
