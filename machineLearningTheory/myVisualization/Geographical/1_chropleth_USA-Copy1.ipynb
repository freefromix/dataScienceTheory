```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import plotly.graph_objs as go
import pandas as pd
```

```python
# Plotly: Generating Offline Graphs within Jupyter Notebook
# You can also plot your graphs offline inside a Jupyter Notebook Environment. First you need to initiate the Plotly Notebook mode as below.
# connected (default=False) -- If True, the plotly.js library will be loaded from an online CDN.
# If False, the plotly.js library will be loaded locally.
init_notebook_mode(connected=True)
```

```python
# type = "choropleth"
# locations = list of states
# locationmode = 'USA-states'
# colorscale = Either: 'pairs', 'Greys, 'Greens', 'Bluered', 'Hots' etc.
data = dict(type = 'choropleth', locations = ['AZ', 'CA', 'NY'], 
            locationmode = 'USA-states',
            colorscale='Jet',
            text=['text 1', 'text 2', 'text 3'],
            z=[1.0, 2.0, 3.0],
            colorbar={'title': 'Colorbar Title Goes Here'})
```

```python
layout = dict(geo={'scope': 'usa'})
```

```python
data
```

```python
layout = dict(geo={'scope': 'usa'})
```

```python
choromap = go.Figure(data = [data], layout=layout)
```

```python
iplot(choromap)
```

# Another example

```python
df = pd.read_csv('2011_us_ag_exports.csv')
```

```python
df.head()
```

```python
data = dict(type = 'choropleth', 
            locationmode='USA-states',
            locations=df['code'], 
            z = df['total exports'], 
            text = df['state'],
            marker = dict(line = dict(color = 'rgb(12, 12, 12)', width = 2)),
            colorbar = {'title': 'Millions USD'})
```

```python
data
```

```python
layout = dict(title = '2011 US Agricuture Exports by State', 
              geo = dict(scope= 'usa', showlakes = True, lakecolor='rgb(85,173,240)'))
```

```python
layout
```

```python
choromap2 = go.Figure(data = [data], layout=layout)
```

```python
iplot(choromap2)
```
