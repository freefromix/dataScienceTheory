```python
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import plotly.graph_objs as go
import pandas as pd
```

```python
# Plotly: Generating Offline Graphs within Jupyter Notebook
# You can also plot your graphs offline inside a Jupyter Notebook Environment. First you need to initiate the Plotly Notebook mode as below.
# connected (default=False) -- If True, the plotly.js library will be loaded from an online CDN.
# If False, the plotly.js library will be loaded locally.
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('2014_world_gdp_with_codes.csv')
```

```python
df.head()
```

```python
# type = "choropleth"
# locations = list of states
# locationmode = 'USA-states'
# colorscale = Either: 'pairs', 'Greys, 'Greens', 'Bluered', 'Hots' etc.
data = dict(type = 'choropleth',
            locations=df['CODE'],
            colorscale='Jet',
            text=df['COUNTRY'],
            z=df['GDP (BILLIONS)'],
            colorbar={'title': 'GDP in Billions USD '})
```

```python
layout = dict(title='2014 Global GDP',
             geo=dict(showframe=True, projection = {'type': 'equirectangular'}))
```

```python
choromap3 = go.Figure(data=[data], layout=layout)
```

```python
iplot(choromap3)
```
